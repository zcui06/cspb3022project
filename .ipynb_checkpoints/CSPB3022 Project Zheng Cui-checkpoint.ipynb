## 1. Project Description

One of the most crucial material in construction industry is concrete. As concrete is a non-homogenous and anisotropic material, it's phycial properties, such as the conpressive strength, cannot be determined acurately. The compressive strength of concrete, which describes the maximum compression it can take, is a nonlinear and unknown function of various parameters including cement, water, aggregate etc. Currently, most authorities globally stipulate concrete compressive strength with empirical functions that coveres large confidence intervals, which means the strength adopted for structural design could be much less than the real strength. 

Therefore, better prediction of the concrete strength would reduce the gap between the design concrete strength and the real strength, which would improve the structural capacity of the structure and/ or reduce construction cost by choosing permissible but lower grade concrete.

This project will conduct research and optimisation to predict the concrete strength with given predictors (parameters). This will be achieved by regression as the strength of concrete is a continous variable. Although the data from this project were extracted from existing data available online, the topic may have not been fully studied and may require further research.

### 4.2 Cross Validation

Before further optimising the models, we need to address the "Bias - Variance Trade-Off" problem. As we optimise the model, As the value of $R^2$ increases, the model becomes more flexible. However, this may not be a good thing as no matter how well the model performs in training, the model may perform badly in testing. 

K-Fold Cross validation is a good way to addres this issue. The general procedure is as follows:
1. Divide the 80% of the data to training data and the rest 20% to be the final testing data. 
2. Evenly split the training data into k groups (k=5 for this project)
3. Take Group 1 as the training data and the rest as training data.
4. Take Group 2 as the training data and the rest as training data.
5. Take Group 3 as the training data and the rest as training data.
6. Take Group 4 as the training data and the rest as training data.
7. Take Group 5 as the training data and the rest as training data.
8. Summarise and analyse the evaluations scores ($R^2$)

The above steps are realised by patsy and sklearn packages in Python.

```python
from IPython.display import Image
Image("grid_search_cross_validation.png")
```

```python
from sklearn import datasets
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn import linear_model,metrics
from sklearn.utils import shuffle
import patsy
```

```python
df_rand = shuffle(df) # shuffle the data frame
```

```python
num_rows = len(df_rand.index) # get the number of rows in the data frame
df_final_testing = df_rand[0:num_rows//5] # 20% of final testing data
df_training = df_rand[num_rows//5:] # 80% of training data

# training matricies
y, X = patsy.dmatrices('strength~cement+slag+ash+water+spp+coarseAgg+fineAgg+age', data=df_training)
# final testing matricies
y_test, X_test = patsy.dmatrices('strength~cement+slag+ash+water+spp+coarseAgg+fineAgg+age', data=df_final_testing)
```

```python
y
```

```python
X
```

```python
df_KFold_linear = linear_model.LinearRegression()
df_KFold_linear_predicted = cross_val_predict(df_KFold_linear, X_test, y_test, cv=5) # predictions for the final tests
scores = cross_val_score(df_KFold_linear, X, y, cv=5, scoring='r2')
print("Cross validation R^2 scores are:", scores)
print("Average Cross validation R^2 score is:", scores.mean())
print("Compare with R^2 in the previous model without cross validation:",lin_model.rsquared_adj)
```

**We will be checking the model with cross validation as we further optimise the models**

### 4.2 Model Optimisation - Backward Stepwise Method

**Next we use Backward Stepwise Method to check if we can reduce the $R^2$ values. We start with removing one predictor from the model**

```python
# list of allowed factors can be removed
allowed_factors=list(df_training.columns)
allowed_factors=allowed_factors[0:len(allowed_factors)-1]

rsquaredadj_lst_bw1 = [] # initialise the list to store the R^2 value for each new model

# create the strings of all possible models and calculate the R^2 adjusted
formula_lst=[]
for i in range(len(allowed_factors)):
    formula = "strength~"
    for j in range(len(allowed_factors)):
        if i != j:
            formula = formula + allowed_factors[j] + "+"
    formula = formula[:-1]
    formula_lst.append(formula)
    lin_model_bw1 = smf.ols(formula=formula, data=df_rand).fit()
    rsquaredadj_lst_bw1.append(lin_model_bw1.rsquared_adj)
```

```python
formula_lst
```

```python
rsquaredadj_lst_bw1
```

```python
# the best solution after one step of backward method
idx=rsquaredadj_lst_bw1.index(max(rsquaredadj_lst_bw1))
formula = formula_lst[idx]
print("The best solution is", formula,"with adjusted R Squared =", max(rsquaredadj_lst_bw1))
```

**With Adjusted $R^2$=0.612 for the full linear model, there is minimum change to the backward method. This is ideal because we are targeting a model with least number of predictors with a reasonably high $R^2$ value**

**Let's double check the cross valudated $R^2$ scores**

```python
# create matricies for training and final testing data
y, X = patsy.dmatrices(formula, data=df_training)
y_test, X_test = patsy.dmatrices(formula, data=df_final_testing)

# create an instance of linear model
df_KFold_linear = linear_model.LinearRegression()
# k-fold cross validation with k=5. the returned object contains the predicted values for the final testing data
df_KFold_linear_predicted_bw1 = cross_val_predict(df_KFold_linear, X_test, y_test, cv=5)
# k-fold cross validation with k=5. the returned object contains the r2 values during traning
scores_bw1 = cross_val_score(df_KFold_linear, X, y, cv=5, scoring='r2')
print("Cross validation R^2 scores are:", scores_bw1)
print("Average Cross validation R^2 score is:", scores_bw1.mean())
```

**Now let's carry on the backward method and try to remove one more predictor.**

```python
# list of allowed factors can be removed. fineAgg is removed because it was dropped in the previous step.
allowed_factors.remove("fineAgg")

rsquaredadj_lst_bw2 = [] # initialise the list to store the R^2 value for each new model

# create the strings of all possible models and calculate the R^2 adjusted
formula_lst=[]
for i in range(len(allowed_factors)):
    formula = "strength~"
    for j in range(len(allowed_factors)):
        if i != j:
            formula = formula + allowed_factors[j] + "+"
    formula = formula[:-1]
    formula_lst.append(formula)
    lin_model_bw2 = smf.ols(formula=formula, data=df_rand).fit()
    rsquaredadj_lst_bw2.append(lin_model_bw2.rsquared_adj)
```

```python
formula_lst
```

```python
rsquaredadj_lst_bw2
```

```python
# the best solution after two steps of backward method
idx=rsquaredadj_lst_bw2.index(max(rsquaredadj_lst_bw2))
formula = formula_lst[idx]
print("The best solution is", formula,"with adjusted R Squared =", max(rsquaredadj_lst_bw2))
```

```python
# create matricies for training and final testing data
y, X = patsy.dmatrices(formula, data=df_training)
y_test, X_test = patsy.dmatrices(formula, data=df_final_testing)

# create an instance of linear model
df_KFold_linear = linear_model.LinearRegression()
# k-fold cross validation with k=5. the returned object contains the predicted values for the final testing data
df_KFold_linear_predicted_bw2 = cross_val_predict(df_KFold_linear, X_test, y_test, cv=5)
# k-fold cross validation with k=5. the returned object contains the r2 values during traning
scores_bw2 = cross_val_score(df_KFold_linear, X, y, cv=5, scoring='r2')
print("Cross validation R^2 scores are:", scores_bw2)
print("Average Cross validation R^2 score is:", scores_bw2.mean())
```

**In summary, the backward stepwise method could remove 2 predictors, coarseAgg and fineAgg from the full linear model while maintaining a close fit model. This makes sense as although the coarse and fine aggregates are crucial and major components of concrete, taking large proportion means they may be less sensitive and less influential to the concrete strength with small changes. In addition, the mixture of all the other materials act as an adhesive that contributes more to the concrete strength.**

```python
x = ["Full Linear", "Backward -1", "Backward -2"]
y = [lin_model.rsquared_adj, max(rsquaredadj_lst_bw1), max(rsquaredadj_lst_bw2)]
y_cv = [scores.mean(), scores_bw1.mean(), scores_bw2.mean()]
print(y)
print(y_cv)
plt.plot(x, y)
plt.plot(x, y_cv)
plt.xticks(x)
plt.title('R-squared scores');
```

```python
plt.plot(df_final_testing.strength, df_KFold_linear_predicted_bw2, 'ro');
plt.plot([df_final_testing.strength.min(), df_final_testing.strength.max()], [df_final_testing.strength.min(), df_final_testing.strength.max()], 'k--', lw=4)
plt.xlabel('Measured')
plt.ylabel('Predicted');
plt.title('Measured Strength VS Predicted Strength');
```

### 4.4 Model Optimisation - Engineering Experience

**As mentioned in the EDA section, Water has higher correlation with some other concrete components than other component pairs. In researches in concrete materials, the term "water-to-cement ratio" has major effect to the concrete structural microscopically. We will further reseach if and how the interaction between these two predictors affect the concrete strength.**

```python
from IPython.display import Image
Image("wc ratio.jpg")
```

```python
inter_model_wc = smf.ols(formula='strength~cement*water + cement*spp + \
                      slag + \
                      ash*water + ash*spp + \
                      water*spp + \
                      spp*age + np.divide(water, cement)', data=df).fit()
inter_model_wc.summary()
```

**The table shows that the water to cement rationot only  has a low P value, but also have a high coefficient value, making it an important predictor to predict the concrete strength**

**Check predictions with cross validation**

```python
formula='strength~cement*water + cement*spp + \
                      slag + \
                      ash*water + ash*spp + \
                      water*spp + \
                      spp*age + np.divide(water, cement)'
# create matricies for training and final testing data
y, X = patsy.dmatrices(formula, data=df_training)
y_test, X_test = patsy.dmatrices(formula, data=df_final_testing)

# create an instance of linear model
df_KFold_linear = linear_model.LinearRegression()
# k-fold cross validation with k=5. the returned object contains the predicted values for the final testing data
df_KFold_linear_predicted_wc = cross_val_predict(df_KFold_linear, X_test, y_test, cv=5)
# k-fold cross validation with k=5. the returned object contains the r2 values during traning
scores_wc = cross_val_score(df_KFold_linear, X, y, cv=5, scoring='r2')
print("Cross validation R^2 scores are:", scores_wc)
print("Average Cross validation R^2 score is:", scores_wc.mean())
```

## 5. Conclusion and Future Improvements

Although there concrete is not a homogeneous material, we have achieved an optimised model that can predict the concrete strength. It started from a simple multilinear model, then was simplified by backward method. Furture optimisations were made through adding interaction terms. The final model the project has achieved is:

$$Strength = -51.7 + 0.32\: Cement + 0.25\: Water - 1.24\: Superplasticiser + 0.08\: Slag + 0.45\:  Ash + 0.09\: Age - 0.001\: Cement\times Water - 0.004\: Cement\times Superplasticiser - 0.002\:  Ash\times Water - 0.008\:  Ash\times Superplasticier + 0.014\:  Water\times Superplasticier + 0.02\:  Superplasticier\times Age - 7.55\:  \frac{Water}{Cement}$$

One recommendation of future improvement to the model is to future investigate the role of coarse and fine aggregates in concrete strength. These two predictors were dropped due to the low coeffcients and P values in the learning results. This low sensitivity was discuss to be likely due to the large proportion of aggregates in the concrete proposition. Further studies could find measurments to scale down the weights of aggregates in the model to reveal the influence of aggregate quantities to the concrete strength.
